# Lecture 3: Tensor Operations

## Section 1: Reshape and View
**In summary differences between shape and view:**
1. Use .view() when you are certain the tensor is contiguous, and you desire memory sharing (i.e., you want changes to reflect in both the original and reshaped tensor).
2. Use .reshape() when you need to change the shape regardless of contiguity and are less concerned about strict memory sharing, as it might create a copy. If you are unsure about the contiguity of your tensor, reshape() is generally the safer choice as it will not raise an error.

```python
my_torch1 = torch.arange(10)

my_torch1, my_torch1.reshape([2, 5]), my_torch1.view([2, 5])
```

## Section 2: Reshape if we don't know the number of items using -1

```python
my_torch2 = torch.arange(15)

my_torch2.reshape([3, -1]), my_torch2.reshape([-1, 3])
```

## Section 3: With reshape and view, relative tensors will update if original tensors make changes.
**_Also apply to NumPy, Pandas_**

```python
my_torch3 = torch.arange(10)
my_torch4 = my_torch3.reshape([2, 5])

my_torch3, my_torch4
```

```python
my_torch3[1] = 123

my_torch3, my_torch4
```

## Section 4: Slices
1. Grab a Specific Item:

```python
my_torch5 = torch.arange(15)
my_torch5[7]
```

2. Grab Slice:
    * Return in a row, cause [:, 1] mean pick all row and pick column idx=1 item
    * Return in a column, cause [:, 1:2] mean pick all row and pick column form idx=1 to idx=1, number after colon will not be included.

```python
my_torch6 = my_torch5.reshape([5, 3])
my_torch6[:, 1], my_torch6[:, 1:2]
```

# Lecture 4: Tensor Math Operations

**Tensor Math Operations:**
* Add, Subtract, Multiply, Divide, Remainders, Exponents
* Shorthand, Longhand
* Reassignment

```python
a = torch.tensor([1, 2, 3, 4])
b = torch.tensor([5, 6, 7, 8])
```

## Section 1: Addition

```python
a + b, torch.add(a, b), a.add(b)
```

## Section 2: Subtraction

```python
a - b, torch.subtract(a, b)
```

## Section 3: Multiplication

```python
a * b, torch.multiply(a, b)
```

## Section 4: Division

```python
a / b, torch.div(a, b)
```

## Section 5: Remainders Module

```python
b % a, torch.remainder(b, a)
```

## Section 6: Exponents / Power

```python
a ** b, torch.pow(a, b)
```

## Section 7: Reassignment with underscore

**Recommended:**

```python
a = a + b
a
```

```python
a += b
a
```

**Not Recommended:**

```python
a = a.add_(b)
a
```
